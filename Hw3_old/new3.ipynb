```python
def Forward(X,W):
    S = np.matmul(W.T, X)
    X = np.tanh(S)
    return X, S
```

```python
def forwardPropagation(X0,W):
    X_array = [] 
    S_array = []
    X_array.append(X0)
    layer_count = len(W)
    for l in range (0, layer_count):
        if(l == 0):
            X,S = Forward(X0,W[l])
            X = np.vstack([[1], X])
            
        elif(l ==layer_count-1):
            X,S = Forward(X_array[l],W[l])
            
        else:
            X,S = Forward(X_array[l],W[l])
            X = np.vstack([[1], X])
            
        X_array.append(X)
        S_array.append(S)
    return X_array,S_array
```

```python
def tanh_prime(X):
    result = 1- X**2
    return result
```

```python
def pointwise_theta(X):
    result = 1 - np.multiply(X,X)
    return result
```

```python
def pointwise(X,Y):
    result = np.multiply(X,Y)
    return result
```
