```python
def initialWeigths(NEXT_LAYER_SIZE, INPUT_LAYER_SIZE):
    W = np.random.randn(INPUT_LAYER_SIZE, NEXT_LAYER_SIZE) * \
                np.sqrt(2.0/INPUT_LAYER_SIZE)*0.01
    return W
```
