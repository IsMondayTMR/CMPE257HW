```python
#SGD
from sklearn.preprocessing import LabelBinarizer#标签二值化
from sklearn.model_selection import train_test_split#数据集分割
import numpy as np
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
 
def tanh(x):  
    return np.tanh(x)  

def tanh_derivate(x):  
    return 1-np.tanh(x)*np.tanh(x)

class NeuralNetwork:
    def __init__(self,layers):#这里是三层网络，列表[64,100,10]表示输入，隐藏，输出层的单元个数
        #初始化权值，范围1~-1
        self.V=np.random.random((layers[0]+1,layers[1]))*2-1#隐藏层权值(65,100)，之所以是65，因为有偏置W0
        self.W=np.random.random((layers[1],layers[2]))*2-1#(100,10)
        
    def train(self,X,y,lr=0.1,epochs=10000):
	#lr为学习率，epochs为迭代的次数
        #为数据集添加偏置
    
        temp=np.ones([X.shape[0],X.shape[1]+1])
        temp[:,0:-1]=X
        X=temp#这里最后一列为偏置
        #绘制曲线
        epoch_plt = []
        accuracy_plt = []
        #进行权值训练更新
        for n in range(epochs+1):
            #SGD
            i=np.random.randint(X.shape[0])#随机选取一行数据(一个样本)进行更新
            x=X[i]
            x=np.atleast_2d(x)#转为二维数据
            #前馈网络
            L1=tanh(np.dot(x,self.V))#隐层输出(1,100)
            L2=tanh(np.dot(L1,self.W))#输出层输出(1,10)
            
            #反向传播
            #delta
            L2_delta=(y[i]-L2)*tanh_derivate(L2)#(1,10)
            L1_delta=L2_delta.dot(self.W.T)*tanh_derivate(L1)#(1,100)，这里是数组的乘法，对应元素相乘
            
            #误差更新
            self.W+=lr*L1.T.dot(L2_delta)#(100,10)
            self.V+=lr*x.T.dot(L1_delta)#
            
            #每训练1000次预测准确率
            if n%1000==0: 
                predictions=[]
                for j in range(X_test.shape[0]):
                    out=self.predict(X_test[j])#用验证集去测试
                    predictions.append(np.argmax(out))#返回预测结果
                accuracy=np.mean(np.equal(predictions,y_test))#求平均值
                #采用mse作为error分析
                #若用rmse就sqrt开方即可
                error=mean_squared_error(predictions,y_test)
                rmse=math.sqrt(error)
                print('epoch:',n,'error:',rmse)
                epoch_plt.append(n)
                accuracy_plt.append(error)
        #绘制曲线        
        plt.plot(epoch_plt, accuracy_plt, lw=1)
        plt.xlim([0, 2000000])
        #plt.ylim([0, 1.05])
        plt.xlabel("epochs")
        plt.ylabel("error")
        plt.show()
    def predict(self,x):
        #添加转置,这里是一维的
        temp=np.ones(x.shape[0]+1)
        temp[0:-1]=x
        x=temp
        x=np.atleast_2d(x)
        
        L1=tanh(np.dot(x,self.V))#隐层输出
        L2=tanh(np.dot(L1,self.W))#输出层输出
        return L2
 
 

#数据归一化,一般是x=(x-x.min)/x.max-x.min
# X-=X.min()
# X/=X.max()
 
#创建神经网络
nm=NeuralNetwork([256,10,10])
 
```
