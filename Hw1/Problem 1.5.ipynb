```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# define the figure size
plt.rcParams['figure.figsize'] = (12.0, 10.0) 
```

```python

Dataset_size = 100


Dataset = np.random.rand(Dataset_size,2) #generate a set of data with rows of Size and column of 2

x = np.linspace(0,1); # Return evenly spaced numbers over a specified interval.and the number is default to 50

#define the function f
a = np.random.rand()
b = np.random.rand()
c = np.random.rand()
if(c > 0.5):
    function = lambda x: -a*x + b 
else:
    function = lambda x: a*x + b 


figure =plt.figure() #create a new figure
figure_gca = plt.gca(); #Get the current Axes instance on the current figure matching the given keyword args, or create one.

 
plt.plot(x,function(x),'g')

Y_Dataset = np.zeros([Dataset_size,1]) #generate Y array with N rows and 1 column with value 0
plt.plot(Dataset[:,0],Dataset[:,1],'bo')

for i in range(Dataset_size):
    if(function(Dataset[i,0]) > Dataset[i,1]):
        Y_Dataset[i] = -1
        plt.plot(Dataset[i,0],Dataset[i,1],'ro')
    else:
        Y_Dataset[i] = 1;
        
        
plt.legend(['Separator','Above','Below'],loc=1)
plt.title('linear separable data figure with the correct line.')



```

```python
def perceptron(dataset,yn,eta =100,weights=np.zeros(3),):

    
    size = dataset.shape[0];
    count = 0
    error = 0
    # Separating curve
#     f = lambda x: np.sign(weights[0]+weights[1]*x[0]+ weights[2]*x[1]);
    while count<=1000: 
        for _ in range(size):
#             st = np.sign(weights[0]+weights[1]*dataset[i,0]+ weights[2]*dataset[i,1])
            i = np.random.randint(size);
            st = np.sign(weights[0]+weights[1]*dataset[i,0]+ weights[2]*dataset[i,1])
            checker = (st*yn[i])
            if( checker <= 1): # If not classified correctly, adjust the line to account for that point.
                count+=1
                error+=1
                weights[0] = weights[0] + eta*(yn[i]-st)
                weights[1] = weights[1] + eta*(yn[i]-st)*dataset[i,0];
                weights[2] = weights[2] + eta*(yn[i]-st)*dataset[i,1];
#                 weights[0] = weights[0] + yn[i];
#                 weights[1:] = weights[1:] + yn[i]*dataset[i,:];
#                 weights[1] = weights[1] + yn[i]*dataset[i,0];
#                 weights[2] = weights[2] + yn[i]*dataset[i,1];
            
    return weights;
```

```python
w = perceptron(Dataset,Y_Dataset)

slope = -w[1]/w[2];
intercept = -w[0]/w[2];

y = lambda x: slope * x + intercept;
sep_color = (Y_Dataset+1)/2.0;

plt.figure();
figa = plt.gca()

training_error_count = 0;

for i in range(Dataset_size):
    if(y(Dataset[i,0]) > Dataset[i,1] and Y_Dataset[i] == 1):
        training_error_count += 1
    elif(y(Dataset[i,0]) < Dataset[i,1] and Y_Dataset[i] == -1):
        training_error_count += 1
    else:
        training_error_count = training_error_count

training_error_rate = (training_error_count/Dataset_size)*100
print("use the line from PLA compare with trainingdataset, there are {} errors , error rate is {}%".format(training_error_count,training_error_rate))

plt.scatter(Dataset[:,0],Dataset[:,1],c=sep_color, s=30)
plt.plot(x,y(x),'b--',label='PLA.')
plt.plot(x,function(x),'r',label='Original line.')
plt.legend()

plt.title('Comparison')
```
