## (a)

   generate a linearly separable data set of size 20 as indicated in Excercise 1.4. Plot the examples{(xn,yn)} as well as the target function *f* on a plane. Be sure to mark the examples from different classes differently and add labels to the axes of the plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# define the figure size
plt.rcParams['figure.figsize'] = (12.0, 10.0) 
```

```python
# generate point and set the function f

Size = 20 #define the number of vector that will be generated

Dataset = np.random.rand(Size,2) #generate a set of data with rows of Size and column of 2
x = np.linspace(0,1); # Return evenly spaced numbers over a specified interval.and the number is default to 50

#define the function f
a = np.random.rand()
b = np.random.rand()
c = np.random.rand()
if(c > 0.5):
    function = lambda x: -a*x + b 
else:
    function = lambda x: a*x + b 


figure =plt.figure() #create a new figure
figure_gca = plt.gca(); #Get the current Axes instance on the current figure matching the given keyword args, or create one.

 
plt.plot(x,function(x),'g')
plt.plot(Dataset[:,0],Dataset[:,1],'bo');
Y = np.zeros([Size,1]) #generate Y array with N rows and 1 column with value 0

for i in range(Size):
    if(function(Dataset[i,0]) > Dataset[i,1]):
        Y[i] = -1
        plt.plot(Dataset[i,0],Dataset[i,1],'ro')
    else:
        Y[i] = 1;

        
plt.legend(['Separator','Above','below'],loc=1)
plt.title('linear separable data figure with the correct line.')
```

## (b)
   Run the perceptron learning algorithm on the data set above. Report the number of updates that the algorithm takes before converging. Plot the example{(xn,yn)}, the target function *f* and the final hypothesis *g* in the same figure. Comment on whether *f* is close to *g*

```python
def perceptron(dataset,yn,MaxIter=1,weights=np.zeros(3)):

    
    size = dataset.shape[0];
    count = 0
    # Separating curve
    f = lambda x: np.sign(weights[0]+weights[1]*x[0]+weights[2]*x[1]);
    for _ in range(MaxIter):
        for _ in range(size):
            i = np.random.randint(size);
            if(yn[i] != f(dataset[i,:])): # If not classified correctly, adjust the line to account for that point.
                count+=1
                weights[0] = weights[0] + yn[i];
                weights[1:] = weights[1:] + yn[i]*dataset[i,:];
#               weights[1] = weights[1] + yn[i]*dataset[i,0];
#                 weights[2] = weights[2] + yn[i]*dataset[i,1];
    print("update {} times".format(count))
            
    return weights;
```

```python
w = perceptron(Dataset,Y)

slope = -w[1]/w[2];
intercept = -w[0]/w[2];

y = lambda x: slope * x + intercept;
sep_color = (Y+1)/2.0;

plt.figure();
figa = plt.gca()

plt.scatter(Dataset[:,0],Dataset[:,1],c=sep_color, s=30)
plt.plot(x,y(x),'b--',label='PLA.')
plt.plot(x,function(x),'r',label='Original line.')
plt.legend()

plt.title('Comparison')
```

from the graph above, it is obviously that the line generated by PLA is not that close to the original line, and the update times is 6 which means in the 20 iterations, the algorithm found 6 misclassified vectors and updated them
